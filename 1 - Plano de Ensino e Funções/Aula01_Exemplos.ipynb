## Exemplo 01
Plotar o gráfico da função $f(x) = x^2$

```python
x = symbols('x')

plt01 = plot(x**2, show = False)

plt01.xlabel = "x"
plt01.ylabel = "f(x) = x^2"
plt01.show()
```

## Exemplo 05
Achar o grau dos polinômios abaixos:

$ f(x) = x + 3 $

$ g(x) = 3x^2 + 4x + 1 $

$ h(x) = 5x^3 + 10x^2 - 8x - 7 $

```python
# Grau de f(x)
f = x + 3
degree(f)
```

```python
x = symbols('x')

plt = plot(f, show = False)

plt.xlabel = "x"
plt.ylabel = "f(x) = " + str(f)

plt.show()
```

```python
# Grau de g(x)
g = 3*x**2 + 4*x + 1
degree(g)
```

```python
x = symbols('x')

plt = plot(g, show = False)

plt.xlabel = "x"
plt.ylabel = "g(x) = " + str(g)

plt.xlim = (-3, 3)
plt.ylim = (-3, 3)

plt.show()
```

```python
# Grau de h(x)
h = 5*x**3 + 10*x**2 - 8*x - 7
degree(h)
```

```python
x = symbols('x')

plt = plot(h, show = False)

plt.xlabel = "x"
plt.ylabel = "f(x) = " + str(h)

plt.show()
```

## Exemplo 06
Operações com polinômios

## Soma
$$ 
(-7x^3 + 5x^2 -x + 4) + (-2x^2 + 8x - 7)
$$

```python
eq01 = -7*x**3 + 5*x**2 - x + 4

eq02 = -2*x**2 + 8*x - 7

res = eq01 + eq02
res
```

## Subtração
$$
(4x^2 - 5x + 6) - (3x - 8)
$$

```python
eq01 = 4*x**2 - 5*x + 6

eq02 = 3*x - 8

res = eq01 - eq02
res
```

## Multiplicação
$$
(3x^2 - 5x + 8) \times (-2x + 1)
$$

```python
eq01 = 3*x**2 - 5*x + 8

eq02 = -2*x + 1

res = eq01 * eq02
res.simplify()
```

## Divisão
$$
\frac{3x^3 - 14x^2 + 23x + 10}{x^2 - 4x + 5}
$$

```python
eq01 = 3*x**3 - 14*x**2 + 23*x - 10

eq02 = x**2 - 4*x + 5
```

```python
res = eq01/eq02
res
```

```python
res = pdiv(eq01, eq02)
res
```
