# 1º QUESTÃO

<font size = 5>

$$
f(x) = \frac{-2}{\sqrt{x-3}}
$$

</font>

```python
f = -2/(sqrt(x-3))
f
```

```python
# Plotando a função 'f'
plt = plot(f, (x, 2, 10), show = False)
plt.ylim = [-10, 0]

plt.show()
```

```python
# Verificando se tem assíntota vertical em x = 3
limit(f, x, 3, '+')
```

```python
# Verificando se tem assíntota horizontal
limit(f, x, oo)
```

```python
# Plotando a função 'f'
plt = plot(f, (x, 2, 10), show = False)

# Plotando a  assíntota vertical em x = 3
plt1 = plot_implicit(Eq(x, 3), (x, 2, 4), (y, -10, 0), line_color='#de1fde', show = False, adaptive = False)
# Plotando a assíntota horizontal em y = 0
plt2 = plot(0, (x, 2, 10), line_color = 'g', show = False)

plt.append(plt1[0])
plt.append(plt2[0])
plt.ylim = [-10, 0]

plt.show()
```

# 2º QUESTÃO

<font size = 5>

$$
f(x) = \frac{x^2}{e^x}
$$

</font>

```python
f = (x**2)/(exp(x))
f
```

```python
# Plotando a função 'f'
plt = plot(f, (x, -5, 5))
```

```python
# Verificando se tem assíntota vertical em x = 5
limit(f, x, -5, '+')

# Como o resultado não é infinito, logo não é uma assíntota vertical
```

```python
# Verificando se tem assíntota horizontal
limit(f, x, 00)
```

```python
# Plote da função f
plt = plot(f, (x, -5, 5), show = False)

# Plote da assíntota vertical em y = 0
plt1 = plot(0, (x, -5, 5), line_color = 'g', show = False)

plt.append(plt1[0])

plt.show()
```

# 4º QUESTÃO

<font size = 5>

$$
f(x) = \frac{x^3 +2x^2 -3}{5x^2+1}
$$

</font>

```python
f = (x**3 + 2*x**2 - 3)/(5*x**2 + 1)
f
```

```python
# Plot da função 'f'
plt = plot(f, (x, -10, 10))
```

```python
# Verificando se tem assíntota vertical em x = 0
limit(f, x, 0, '-')

# Como o resultado não é infinito, logo não é uma assíntota
```

```python
# Verificando se tem assíntota horizontal
limit(f, x, oo)

# Como o resultado não é um valor definido, logo não é uma assíntota horizontal
```

```python
# Verificando se tem assíntota horizontal
limit(f, x, -oo)

# Como o resultado não é um valor definido, logo não é uma assíntota horizontal
```

```python
# Calculando o valor de 'f' para x = 10000
N(f.subs(x, 10000))
```

```python
# Verificando a proporção de 'y' com 'x' para achar o valor da inclinação da reta
2000.399/10000
```

```python
# Calculando o limite para ver se a reta 'y = 0.2x' é assíntota
limit(f - (0.2*x), x, oo)

# O resultado para ter um valor definido significa que acertamos a angulação.
# Porém como o resultado é diferente de 0, que dizer que não estamos com o valor de 'b' correto.
```

```python
# Calculando o limite para ver se a reta 'y = 0.2x + 2/5' é assíntota
limit(f - (0.2*x + 2/5), x, oo)

# A reta 'y = 0.2x + 2/5' é assintota diagonal
```

```python
# Plot da função 'f'
plt = plot(f, (x, -10, 10), show = False)

# Plot da assíntota diagonal
plt1 = plot(0.2*x + 0.4, (x, -10, 10), line_color = 'g', show = False)

plt.append(plt1[0])

plt.show()
```
