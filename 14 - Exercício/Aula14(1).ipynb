# 2º QUESTÃO

<font size = 5>

Verificar as assíntotas horizontais e verticais da função abaixo.

$$
f(x) = \frac{-1}{(x-3)(x+4)}
$$

</font>

```python
f = -1/((x-3)*(x+4))
f
```

```python
plt = plot(f, (x, -5, 5), show = False)

plt.ylim = [-10, 10]

plt.show()
```

```python
limit(f, x, -4, '+')
```

```python
limit(f, x, -4, '-')
```

```python
limit(f, x, 3, '+')
```

```python
limit(f, x, 3, '-')
```

```python
limit(f, x, oo)
```

```python
limit(f, x, -oo)
```

```python
plt = plot(f, (x, -5, 5), show = False)

plt1 = plot_implicit(Eq(x, -4), (x, -3, -2), (y, -10, 10), show = False, line_color = 'r', adaptive = False)
plt2 = plot_implicit(Eq(x, 3), (x, 2, 4), (y, -10, 10), show = False, line_color = 'r', adaptive = False)

plt3 = plot(0, (x, -5, 5), line_color = 'y', show = False)

plt.append(plt1[0])
plt.append(plt2[0])
plt.append(plt3[0])

plt.ylim = [-10, 10]
plt.show()
```
