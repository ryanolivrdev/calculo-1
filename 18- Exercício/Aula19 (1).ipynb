Em que pontos o gr√°fico da fun√ß√£o $y = \frac{1}{3}x^3 - \frac{2}{3}x^2 + 3x$ tem tangente horizontal?

Plote o gr√°fico e analisar o resultado obtido

</font>

```python
f = Rational(1, 3)*x**3 - Rational(3/2)*x**2 + 2*x
f
```

```python
plt = plot(f, (x,0, 3), show = False)
plt.title = "Gr√°fico da fun√ß√£o f"

plt.show()
```

<font size = 5>

Para descobrir onde uma reta tangete se comporta como uma reta horizontal, basta descobrir os pontos iguais a zero na derivada da fun√ß√£o

</font>

```python
df = f.diff(x)
df
```

```python
plt = plot(df, (x, 0, 3), show = False)
plt.title = "√Çngulo da reta tangente"

plt.show()
```

<font size = 5>

Tudo indica pelo gr√°fico que os pontos s√£o $x = 1$ e $x = 2$

</font>

```python
# Resolvendo a equa√ß√£o df para achar onde o 'x' √© igual a zero.
solve(df, x)
```
