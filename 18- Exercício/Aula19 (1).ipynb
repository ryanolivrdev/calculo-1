# 1º Questão

<font size = 5>

Em que pontos o gráfico da função $y = \frac{1}{3}x^3 - \frac{2}{3}x^2 + 3x$ tem tangente horizontal?

Plote o gráfico e analisar o resultado obtido

</font>

```python
f = Rational(1, 3)*x**3 - Rational(3/2)*x**2 + 2*x
f
```

```python
plt = plot(f, (x,0, 3), show = False)
plt.title = "Gráfico da função f"

plt.show()
```

<font size = 5>

Para descobrir onde uma reta tangete se comporta como uma reta horizontal, basta descobrir os pontos iguais a zero na derivada da função

</font>

```python
df = f.diff(x)
df
```

```python
plt = plot(df, (x, 0, 3), show = False)
plt.title = "Ângulo da reta tangente"

plt.show()
```

<font size = 5>

Tudo indica pelo gráfico que os pontos são $x = 1$ e $x = 2$

</font>

```python
# Resolvendo a equação df para achar onde o 'x' é igual a zero.
solve(df, x)
```

<font size = 5>

Um corpo cai livremente partindo do repouso. Calcule sua posição e sua velocidade depois de decorridos 1 e 2 segundos. Lembrando que a equação utilizada na física é $𝒚 = 𝒗_𝒐𝒕 − \frac{1}{2}𝒈𝒕^𝟐$.

Como o corpor está partindo do repouso, então $v_0 = 0$.

</font>

```python
# Função do espaço
f = -(1/2)*9.8*x**2
f
```

```python
# Espaço em t = 1s
f.subs(x, 1)
```

```python
# Espaço em t = 2s
f.subs(x, 2)
```

```python
# Função da velocidade
df = f.diff(x)
df
```

```python
# Velocidade em t = 1s
df.subs(x, 1)
```

```python
# Velocidade em t = 2s
df.subs(x, 2)
```
