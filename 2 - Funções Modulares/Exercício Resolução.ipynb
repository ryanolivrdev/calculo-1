## 1º Questão
Dada a função

$$
f(x) = |x| - 2x
$$

* Calcular $f(-1)$, $f(1/2)$ e $f(-2/3)$
* Mostrar que $f(|a|) = -|a|$

```python
# Criando a função da questão
f = Abs(x) - 2*x
```

```python
# Calculando f(-1)
y = f.subs(x, -1)
y
```

```python
# Calculando f(1/2)
y = f.subs(x, 1/2)
y
```

```python
# Calculando f(-2/3)
y = f.subs(x, -2/3)
y
```

```python
# Calculando f(|a|)
y = f.subs(x, Abs(a))
y
```

## 3º Questão
Um fabricante produz peças para computadores pelo preço de 2,00 reais cada uma. Calcula-se que, se cada peça for vendida por $x$ reais, os consumidores comprarão, por mês, $600-x$ unidades. Expressar o lucro mensal do fabricante como função do preço. Contruir um gráfico para estimar o preço ótimo de venda.

```python
# As unidades compradas de peças por computador é dado pela equação
U = 600 - x

# Multiplicando a quantiadade procurada pelo preço temos a receita
R = U*x

# O custo por cada unidade é de 2 reais cada
C = 2*U

# O lucro é dado pela receita menos o custo
L = R - C
L
```

```python
# Com o lucro, podemos plotar o gráfico e estimar o preço ótimo
plt = plot(L, (x, 0, 600), show = False)

plt.xlabel = 'preço da unidade'
plt.ylabel = 'Lucro'
plt.title = 'Gráfico do lucro mensal dado pelo preço do produto'
plt.show()
```

## 6º Questão
Uma indústria comercializa um certo produto e tem um função custo total dada por $C(x) = x^2 + 20x + 700$, sendo $x$ o número de unidades produzidas. A função receita total é dada por $R(x) = 200x$. Determine:

* O lucro para a venda de 100 unidades.
* Em que valor de x acontecerá o lucro máximo?

```python
# O lucro total é dado pela subtraindo da receita o custo
C = x**2 + 20*x + 700
R = 200*x
L = R - C
L
```

```python
# O lucro para a venda de 100 unidade é dado abaixo
L.subs(x, 100)
```

```python
# Com o lucro, podemos plotar o gráfico e estimar a quantidade de unidades para o lucro máximo
plt = plot(L, (x, 0, 200), show = False)

plt.xlabel = 'unidades'
plt.ylabel = 'Lucro'
plt.title = 'Gráfico do lucro pelo produto'
plt.show()
```

## 7º Questão
A meia-vida do rádio-226 é de 1620 anos.

* Obter o modelo de decaimento exponenccial para essa substância
* Após 700 anos, qual o percentual de uma dada quantidade inicial de rádio ainda resta?

```python
# Modelo do decaimento
M = M0*exp(-x*k)
M
# x é o tempo
# M0 é a massa inicial
# k é uma constante que depende do material
```

```python
# Para x = 1620, a massa é a metade da massa inicial
# Com isso podemos achar o valor de k

K = solve(1/2 - exp(-1620*k), k)
```

```python
# Pegando o valor real
Kr = K[0]
Kr
```

```python
# Então o Modelo do decaimento é
M = M0*exp(-Kr*x)
M
```

```python
# Considerando a massa inicial como 1
M = exp(-Kr*x)

# Criando o objeto do plot
plt = plot(M, (x, 0, 2000), show = False)

# Configuração do objeto
plt.xlabel = 'x'
plt.ylabel = ''
plt[0].line_color = '#0aa3f5'
plt.axis_center = [0,0]
plt.title = M

plt.show()
```

```python
# Se a equação realmente estiver correta, devemos ter a metade do original no tempo 1620
M.subs(x, 1620)
```

```python
# Para 700 anos tempo a porcentagem em relação a massa original
M.subs(x, 700)
```
