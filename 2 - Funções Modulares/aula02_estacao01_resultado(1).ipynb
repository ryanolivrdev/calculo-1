### Pergunta 01
O que é uma função modular?

Lembre-se de colocar as referências de onde você retirou as informações

### Resposta:

Para trabalhar com funções modulares, é importante lembrar a definição de módulo e suas propriedades. Representamos o módulo de um número $n$ por $|n|$, e, por definição, temos que:
$$
|n| = \left\{ \begin{matrix}
n \rightarrow n \geq 0 \\
-n \rightarrow n < 0
\end{matrix} \right.
$$

O módulo de um número sempre gera resultados positivos, por exemplo:
$$
|2| \rightarrow \text{ como } 2 > 0 \rightarrow |2| = 2
$$
$$
|-2| \rightarrow \text{ como } 2 < 0 \rightarrow |-2| = -(-2) = 2
$$

Então, uma função modular é aquela função, com domínio e contradomínio nos números reais, que possui pelo menos uma variável dentro do módulo, $|x|$, em sua lei de formação.

### Pergunta 02
Faça o plote da seguinte função:
$$
f(x) = \left| x \right|
$$

```python
# Vamos comparar a função normal e a função modular
f = x
fm = abs(x)

plt = plot(f, fm, (x, -5, 5), show = False)

plt.title = "Função Modular"
plt.legend = True

plt.show()
```

### Pergunta 03
Qual seria a raiz da função anterior?

Em matemática, uma raiz ou "zero" da função consiste em determinar os pontos de intersecção do gráfico da função com o eixo das abscissas no plano cartesiano.

```python
solveset(abs(x), domain=S.Reals)
```

### Pergunta 04
Faça o plote da seguinte função:
$$
f(x) = \left| x^2 - 6x + 5 \right|
$$

```python
# Vamos comparar a função normal e a função modular
f = x**2 - 6*x + 5
fm = abs(x**2 - 6*x + 5)

plt = plot(f, fm, (x, -2, 8), show = False)

plt.title = "Função Modular"
plt.legend = True

plt.show()
```
