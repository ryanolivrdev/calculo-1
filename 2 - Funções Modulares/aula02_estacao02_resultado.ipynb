### Pergunta 01
O que é uma função exponencial?

Lembre-se de colocar as referências de onde você retirou as informações

### Resposta:
A função exponencial acontece quando a incógnita de um cálculo matemático está no expoente. Dessa forma, seu comportamento gráfico e as formas de resolver dependem do conhecimento a respeito de logaritmos, exponenciação, potenciação, radiciação e etc.

A estrutura geral desse tipo de função será tal que $f(x) = a^x$, de forma que a pertence ao conjunto dos números reais, $a>0$ e $a≠1$.

[Referência Bibliografica](https://vestibulares.estrategia.com/portal/materias/matematica/funcao-exponencial/#:~:text=A%20fun%C3%A7%C3%A3o%20exponencial%20acontece%20quando,%2C%20potencia%C3%A7%C3%A3o%2C%20radicia%C3%A7%C3%A3o%20e%20etc.)

### Pergunta 02
Faça o plote da seguinte função:
$$
f(x) = 2^x
$$

```python
f = 2**x
g = 2.5**x

plt = plot(f, g, (x, -5, 5), show = False)

plt.title = "Função Exponencial"
plt.legend = True

plt.show()
```

### Pergunta 03
Faça o plote da seguinte função:
$$
f(x) = 0.5^x
$$

### Pergunta 04
Olhando para a diferença entre os dois gráficos. O que aconteceu com ele quando o valor da báse foi maior que 1 e quando foi menor que 1?

### Resposta:

Uma função exponencial crescente acontece quando a base é um número real maior do que 1. Isso significa que, quanto maior o valor da incógnita no expoente, maior será o resultado da função. 

Quando a base está entre os valores de 0 e 1 $(0<a<1)$, considera-se que é uma função exponencial decrescente. Na prática, indica que, quanto maior o valor de $x$, menor será o resultado obtido em $y$.
