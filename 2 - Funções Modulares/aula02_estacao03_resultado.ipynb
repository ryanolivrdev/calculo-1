### Pergunta 01
O que é uma função logarítmica?

Lembre-se de colocar as referências de onde você retirou as informações

### Resposta:

Toda função definida pela lei de formação $f(x) = log_a x$, com $a ≠ 1$ e $a > 0$, é denominada função logarítmica de base $a$.

O domínio dessa função está no conjunto dos números reais positivos, sendo diferente de zero, e o contradomínio, no conjunto dos números reais. Vale salientar também que a base do logaritmo tem que ser necessariamente maior que 0 e diferente de 1.

O logaritmo é a operação inversa da potenciação. Sendo assim, a relação entre a função logarítmica e a função exponencial é que aquela é a função inversa desta, e vice-versa.

### Pergunta 02
Faça o plote da seguinte função:
$$
f(x) = \log_3 x
$$

```python
f = log(x, 3)
g = log(x, 4)

plt = plot(f, g, (x, -5, 5), show = False)

plt.title = "Função Logaritmica"
plt.legend = True

plt.show()
```

### Pergunta 04
Olhando para a diferença entre os dois gráficos. O que aconteceu com ele quando o valor da báse foi maior que 1 e quando foi menor que 1?

### Resposta:

Uma função logarítmica crescente acontece quando a base é um número real maior do que 1. Isso significa que, quanto maior o valor da incógnita no expoente, maior será o resultado da função. 

Quando a base está entre os valores de 0 e 1 $(0<a<1)$, considera-se que é uma função exponencial decrescente. Na prática, indica que, quanto maior o valor de $x$, menor será o resultado obtido em $y$.
