```python
from sympy import *
```

```python
x = symbols('x')
y = symbols('y')
```

```python
f = 3*x**2 + 8*x + 1
f
```

```python
plot(f)
```

```python
f1 = f.diff(x)
f1
```

```python
plot(f1)
```

```python
f2 = f1.diff(x)
f2
```

```python
plot(f2)
```

____________________

```python
f = tan(x)
f
```

```python
plt = plot(f, show = False)
plt.ylim = [-10, 10]

plt.show()
```

```python
f1 = f.diff(x)
f1
```

```python
plt = plot(f1, show = False)
plt.ylim = [-10, 10]

plt.show()
```

```python
f2 = f1.diff(x)
f2
```

```python
plt = plot(f2, show = False)
plt.ylim = [-10, 10]

plt.show()
```

```python
f = x**2 + y**2 - 4
f
```

```python
f1 = idiff(x**2 + y**2 - 4, y, x)
f1
```

```python
f = x*y**2 + 2*y**3 - x + 2*y
f
```

```python
f1 = idiff(f, y, x)
f1
```
