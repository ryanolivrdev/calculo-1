# 1¬∞ QUEST√ÉO

<font size = 5>

Dada a fun√ß√£o abaixo.  Mostrar que ùíá satisfaz as condi√ß√µes do Teorema de Rolle no intervalo [-3, 3] e determinar os valores de ùíÑ em (-3,3) que satisfa√ßam ùíá'(ùíÑ) = ùüé.

$$
f(x) = -x^4 + 8x^2 + 9
$$

</font>

```python
f = -x**4 +8*x**2 + 9
f
```

```python
print(f"Valor de f(-3) = {f.subs(x, -3)}")
print(f"Valor de f(3) = {f.subs(x, 3)}")
```

<font size = 5>

Valor de _a_ e _b_ s√£o iguais, logo pelo teorema de Roller a derivada deve ser igual a 0 em pelo menos um ponto.

</font>

```python
df = f.diff(x)
df
```

```python
pontos = solve(Eq(df, 0), x)
for i in pontos:
    print(f"Raiz = {i}")
```

```python
plot(df, (x, -3, 3), title = "Plot Derivada")
```

```python
# reta horizontal em x = -2, 0 e 2
plot(f, (x, -3, 3), title = "Plot Fun√ß√£o")
```
