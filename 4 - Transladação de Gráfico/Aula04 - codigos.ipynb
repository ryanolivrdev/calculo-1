```python
from sympy import *
```

```python
x = symbols('x')
```

## Transladação Horizontal

```python
pltA = plot(x**2, (x,-3,3), show = False)

pltB = plot((x-1)**2, (x, -2, 4), show = False)
pltB[0].line_color = 'r'

pltC = plot((x+1)**2, (x, -4, 2), show = False)
pltC[0].line_color = 'g'

pltA.append(pltB[0])
pltA.append(pltC[0])

pltA.title = "Transladação Horizontal"
pltA.legend = True
pltA.xlim = [-4, 4]

pltA.show()
```

## Transladação Vertical

```python
pltA = plot(x**2, (x,-3,3), show = False)

pltB = plot(x**2 - 1, (x, -3, 3), show = False)
pltB[0].line_color = 'r'

pltC = plot(x**2 + 1, (x, -3, 3), show = False)
pltC[0].line_color = 'g'

pltA.append(pltB[0])
pltA.append(pltC[0])

pltA.title = "Transladação Vertical"
pltA.legend = True
pltA.xlim = [-4, 4]

pltA.show()
```

## Transladação

```python
pltA = plot(x**2, (x,-3,3), show = False)

pltB = plot((x-1)**2 - 1, (x, -2, 4), show = False)
pltB[0].line_color = 'r'

pltC = plot((x+1)**2 + 1, (x, -4, 2), show = False)
pltC[0].line_color = 'g'

pltA.append(pltB[0])
pltA.append(pltC[0])

pltA.title = "Transladação Horizontal e Vertical"
pltA.legend = True
pltA.xlim = [-4, 4]

pltA.show()
```
