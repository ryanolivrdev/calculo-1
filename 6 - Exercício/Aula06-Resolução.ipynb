# 1º Questão

<font size = 5>

Analisar a função $ f(x) = 24x - 3x^2 $ e verificar a possibilidade de representar uma função receita. Em caso afirmativo identifique a função demanda e responda:

    - Qual a quantidade demandada quando o preço unitário é R$ 5,00?
    
    - Qual é o preço do produto quando a receita é máxima?

</font>

```python
# Criando a função receita 
R = 24*x - 3*x**2
R
```

```python
# A função receita tem o formato --> R = p*q
# onde p é igual ao valor do produto
# e q é a quantidade do produto
# Então vamos transformar a função R acima para ter o formato da função receita.

D = 24 - 3*x

R = x*D
R
```

```python
# - Qual a quantidade demandada quando o preço unitário é R$ 5,00?
# D(5) = 24 - 3*5

D.subs(x, 5)
```

```python
# - Qual é o preço do produto quando a receita é máxima?
```

```python
plt = plot(R, (x, 0, 9), show = False)

plt.xlabel = 'Preço (R$)'
plt.ylabel = 'Receita'
plt.title = 'Receita pelo preço do produto'

plt.show()
```

<font size = 5>

Numa dada cidade a população atual é 380.000 habitantes. Se a população apresenta uma taxa de crescimento anual de 1,5%, estime o tempo necessário para a população duplicar.

</font>

```python
# Criando a função exponencial de crescimento populacional.
P = 380000*1.015**x
P
```

```python
# Plotando o gráfico do crescimento populacional
plt = plot(P, (x, 0, 100), show = False)

plt.xlabel = 'ano'
plt.ylabel = 'População'
plt.title = 'Previsão do Crescimento populacional da Cidade X'
plt.axis_center = [0, 0.4*10**6]

plt.show()
```

```python
# O dobro da população inicial é 760.000
# Substituindo na equação temos
# 760.000 = 380.000*1,015^t
# jogando tudo para um lado e usando o solver para achar o resultado

solve(760000 - P, x)
```

# 3º Questão

<font size="5">

Certa empresa fabrica 2 produtos $P_1$ e $P_2$. O lucro por unidade de $P_1$ é de 100 reais e o lucro unitário de $P_2$ é de 150 reais. A empresa necessita de 2 horas para fabricar uma unidade de $P_1$ e 3 horas para fabricar uma unidade de $P_2$. O tempo mensal disponível para essas atividades é de 120 horas. As demandas esperadas para
os dois produtos levaram a empresa a decidir que os montantes produzidos de $P_1$ e $P_2$ não devem ultrapassar 40 unidades de $P_1$ e 30 unidades de $P_2$ por mês. Construa o modelo do sistema de produção mensal com o objetivo de maximizar o lucro da empresa. (Assumir que as quantidades podem ser fracionárias)

</font>

    A ideia com essa questão é escrever todas as equações de restrições e fazer seu plote, para assim, verificar os pontos das extremidades

```python
# Como a produção dos produtos tem um limite de tempo de 120 horas, podemos achar uma relação de P2 com P1.
P2 = (120-2*x)/3
P2
```

```python
# Plotando a equação da restrição do tempo
plt = plot(P2, (x, 0, 60), show = False)

# plotando a reta do limite de produtos P2
plt1 = plot(30, (x, 0, 60), show = False)
plt.append(plt1[0])

# plotando a reta do limite de produtos P1
plt2 = plot_implicit(Eq(x, 40), (x, 0, 60), (y, 0, 40), show = False, line_color = 'r', adaptive = False)
plt.append(plt2[0])

plt.xlabel = 'Produto P1'
plt.ylabel = 'Produto P2'

plt.show()
```

    Analisando o gráfico, temos quatro pontos interessantes:
    > (0, 30)
    > (15, 30)
    > (40, 13.33)
    > (40, 0)

```python
# Criando a equação da receita
L = 100*x + 150*y
```

```python
# Calculando o lucro para P1 = 15 e P2 = 30
L.subs(x, 15).subs(y, 30)
```

```python
# Calculando o lucro para P1 = 40 e P2 = 13.33
L.subs(x, 40).subs(y, 13.333)
```

```python
# Calculando o lucro para P1 = 0 e P2 = 30
L.subs(x, 0).subs(y, 30)
```

```python
# Calculando o lucro para P1 = 40 e P2 = 0
L.subs(x, 40).subs(y, 0)
```

A figura representa um sistema massa-mola ideal, cuja constante elástica é de 4 N/cm.

Um corpo de massa igual a 1,2 kg é empurrado contra a mola, comprimindo-a de 12,0 cm. Ao ser liberado, o corpo desliza ao longo da trajetória representada na figura. Desprezando-se as forças dissipativas em todo o percurso e considerando a aceleração da gravidade igual a 10 m/s², é correto afirmar que a altura máxima H atingida pelo corpo, em cm, é igual a:

</font>

```python
# No ponto máximo a energia elástica é igual a energia potêncial
# Logo podemos fazer uma função da altura que dependa da distância comprimida x

H = 400*(x*10**(-2))**2/(1.2*10*2)
H
```

```python
# Plotando o gráfico da altura
plt = plot(H, (x,0,12), show = False)

plt.xlabel = 'Distancia comprimida da mola (m)'
plt.ylabel = 'Altura máxima da mola (m)'
plt.title = 'Gráfico da altuma máxima da mola pela distância comprimida'

plt.show()
```

```python
# Comprimindo a mola em 12 cm
H.subs(x, 12)
```
