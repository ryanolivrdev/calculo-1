```python
import numpy as np
from sympy import *
```

```python
X = symbols('x')
```

________________

```python
def f(x):
    return 2*x + 1
```

```python
# valores de x tendendo a 1 mas menores que 1
x = np.linspace(0.9, 1, 20)
x = x[:-1]
x
```

```python
print('Atribuindo a x valores próximos de 1 porém menores que 1')

print("x \t | \t f(x)")

for i in x:
    print("{} \t | \t {}".format(round(i, 4), round(f(i), 4)))
```

<font size = 5>

Observemos em ambas as tabelas que, quando *x* se aproxima cada vez mais de 1, *f(x)* aproxima-se cada vez mais de 3.



Portanto, pelas duas tabelas vemos que:

</font>

```python
print('Distância de x para 1')

print("x \t | \t f(x) \t\t | \t |x-1| \t\t | \t |\f(x)-3|")

for i in x:
    print("{} \t | \t {} \t | \t {} \t | \t {}".format(
        round(i, 4), round(f(i), 4), round(abs(i-1), 4), round(abs(f(i)-3), 4)))
```

<font size = 5>

Podemos tornar o módulo da diferença entre f(x) e 3 tão pequeno quanto desejarmos desde que tomemos o módulo da diferença entre x e 1.

</font>

```python
plot(2*X + 1, (X, 0.5, 1.5))
```

```python
limit(2*X + 1, X, 1)
```

```python
def f(x):
    return 1 - 1/x
```

```python
x = np.linspace(1, 1000, 50)
x = x[:-1]
x
```

```python
print('Atribuindo a x valores para o infinito +')

print("x \t | \t f(x)")

for i in x:
    print("{} \t | \t {}".format(round(i, 4), round(f(i), 4)))
```

```python
x = np.linspace(-1, -1000, 50)
x = x[:-1]
x
```

```python
print('Atribuindo a x valores para o infinito -')

print("x \t | \t f(x)")

for i in x:
    print("{} \t | \t {}".format(round(i, 4), round(f(i), 4)))
```

```python
print('Distância de x para infinito (1000)')

print("x \t | \t f(x) \t\t | \t |x-1000| \t\t | \t |\f(x)-1|")

for i in x:
    print("{} \t | \t {} \t | \t {} \t | \t {}".format(
        round(i, 4), round(f(i), 4), round(abs(i-1000), 4), round(abs(f(i)-1), 4)))
```
