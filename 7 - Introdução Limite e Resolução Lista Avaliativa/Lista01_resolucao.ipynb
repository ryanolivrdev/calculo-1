## 2° QUESTÃO

A locadora Alpha aluga um carro popular ao preço de R$ 30,00 a diária mais R$ 0,20 por quilômetro rodado. A locadora Beta o faz por R$ 40,00 a diária mais R$ 0,10 por quilômetro rodado.

Desenvolva um código em Python que faça a plotagem do gráfico do problema e com a análise do gráfico responda a pergunta abaixo:
>   Qual a locadora você escolheria, se você pretendesse alugar um carro por um dia e pagar o menos possível?

```python
# Aqui teremos duas expressões, uma para a locadora Alpha e outra para locadora Beta.
# Olhando primeiro para a locadora Alpha, o preço do aluguel é 30 reais mais 0,2 pela quilometragem rodada.
alpha = 0.2*x + 30
alpha
```

```python
# Olhando agora para a locadora Beta, o preço do aluguel é 40 reais mais 0,1 pela quilometragem rodada.
beta = 0.1*x + 40
beta
```

```python
#plotando o gráfico de alpha
plt = plot(alpha, (x,0, 200), label = 'alpha', show = False)
plt.legend = True

#plotando o gráfico de beta
plt1 = plot(beta, (x,0, 200), label = 'beta', show = False)
plt1.legend = True

plt.append(plt1[0])

plt.title = "Comparação de custo entre as locadoras"
plt.xlabel = "kilometragem rodada"
plt.ylabel = "Custo do aluguel"

plt.show()
```

Analisando o gráfico, verificamos que de 0 até 100 km rodado, a locadora alpha tem um custo menor que a locadora beta, mas esse processo se inverte quando passamos de 100 km rodado

## 3° QUESTÃO

Em um laboratório, um determinado ser vivo apresenta um ciclo produtivo de 1 hora, e a cada hora um par pronto para reprodução gera outro par reprodutor.

Como expressar essa experiência populacional em função do número de horas, supondo que a população inicial é de cinco pares?

Desenvolva o código em Python que possua a função matemática que descreva o comportamento acima e plote seu gráfico.

Aqui estamos trabalhando com crescimento exponencial, para isso podemos fazer um teste rápido:
>   *   $t = 0 \rightarrow p = 5$
>   *   $t = 1 \rightarrow p = 10  \rightarrow 5*2$
>   *   $t = 2 \rightarrow p = 20  \rightarrow [5*2]*2$
>   *   $t = 3 \rightarrow p = 40  \rightarrow [[5*2]*2]*2$
>   *   $t = 4 \rightarrow p = 80  \rightarrow [[[5*2]*2]*2]*2$
>   *   $t = 5 \rightarrow p = 160 \rightarrow [[[[5*2]*2]*2]*2]*2$

onde $t$ presenta o tempo em horas e $p$ o tamanho da população.

Analisando esses dados temos um comportamento exponencial, logo a nossa função pode apresentar a seguinte expressão:

$$
P_t = P_i T^t
$$

Onde $P_t$ é a estimação da populado em um dado tempo; $P_i$ é o tamanho da população inicial; $T$ é a taxa de crescimento e $t$ é o tempo.

Como a população sempre está dobrando de tamanho a cada 1 hora, logo a taxa de crescimento é 2.

A formula final será:

$$
P_t = 5\bullet2^t
$$

```python
# Criando a expressão
p = 5*2**t
p
```

```python
# Fazendo o plote da expressão
plt = plot(p, (t, 0, 10), show = False)
plt.title = "Crescimento de uma população"
plt.xlabel = "Tempo (h)"
plt.ylabel = "Tamanho da população"

plt.show()
```

## 4° QUESTÃO
Dada a função da Figura 02.

$$
f(x) = x^2
$$

Escolha a alternativa que faça o gráfico da função da Figura 02 se deslocar duas unidades para esquerda e cinco unidades para cima

Para andar para a esquerda devemos somar o parâmetro da função com um valor, para esse caso o valor é 2.

Para andar para cima devemos somar a função com um valor, para esse caso o valor é 5.

```python
# Criando a expressão base para fazer a comparação
f1 = x**2
f1
```

```python
# Criando a expressão transladada
f2 = (x+2)**2 + 5
f2
```

```python
# Plotando as duas expressões
plt = plot(f1, (x,-5, 5), show = False)
plt.legend = True

plt1 = plot(f2, (x, -5, 5), show = False)
plt1.legend = True

plt.append(plt1[0])

plt.title = "Comparação entre duas curvas"
plt.xlabel = "x"
plt.ylabel = "f(x)"

plt.show()
```
